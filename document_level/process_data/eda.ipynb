```python
def convert2docred(doc: Dict):
    """Из формата nerel jsons в DocRED
    """
    vertexSet = [
        [
            {
                'name': ment['name'],
                'sent_id': ment['sent_id'],
                'type': ent['type'],
                'pos': ment['token_pos']

            }
            for ment in ent['mentions']
        ]
        for ent in doc['entities']
    ]

    sents = [
        doc['sents'][str(i)]['tokens']
        for i in range(len(doc['sents']))
    ]

    return {
        'title': doc['file_name'],
        'vertexSet': vertexSet,
        'labels': doc['relations'],
        'sents': sents
    }
```

## check max num relation labels for one pair

```python
def max_relation_labels_per_entity_pair(docred_data):
    max_relations = 0
    max_relations_document_idx = -1
    max_relations_entity_pair = None

    for doc_idx, document in enumerate(docred_data):
        entity_pair_relations = {}
        for relation in document['labels']:
            head_idx = relation['h']
            tail_idx = relation['t']
            entity_pair = (head_idx, tail_idx)

            if entity_pair not in entity_pair_relations:
                entity_pair_relations[entity_pair] = 0

            entity_pair_relations[entity_pair] += 1

        current_max_relations = max(entity_pair_relations.values()) if entity_pair_relations else 0
        if current_max_relations > max_relations:
            max_relations = current_max_relations
            max_relations_document_idx = doc_idx
            max_relations_entity_pair = max(entity_pair_relations, key=entity_pair_relations.get)

    return max_relations, max_relations_document_idx, max_relations_entity_pair
```

```python
def max_unique_relation_labels_per_entity_pair(docred_data):
    max_relations = 0
    max_relations_document_idx = -1
    max_relations_entity_pair = None

    for doc_idx, document in enumerate(docred_data):
        entity_pair_relations = {}
        for relation in document['labels']:
            head_idx = relation['h']
            tail_idx = relation['t']
            entity_pair = (head_idx, tail_idx)
            relation_type = relation['r']

            if entity_pair not in entity_pair_relations:
                entity_pair_relations[entity_pair] = set()

            entity_pair_relations[entity_pair].add(relation_type)

        current_max_relations = max(len(relations) for relations in entity_pair_relations.values()) if entity_pair_relations else 0
        if current_max_relations > max_relations:
            max_relations = current_max_relations
            max_relations_document_idx = doc_idx
            max_relations_entity_pair = max(entity_pair_relations, key=lambda pair: len(entity_pair_relations[pair]))

    return max_relations, max_relations_document_idx, max_relations_entity_pair
```
